```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import sklearn
```

```python
breast_dataset= sklearn.datasets.load_breast_cancer()
```

```python
print(breast_dataset)
```

```python
df=pd.DataFrame(breast_dataset.data,columns=breast_dataset.feature_names)
```

```python
df['label']=breast_dataset.target
```

```python
X=df.drop(columns='label',axis=1)
y=df['label']
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_std=scaler.fit_transform(X_train)
X_test_std=scaler.transform(X_test)
```

```python
import tensorflow as tf
```

```python
tf.random.set_seed(3)
```

```python
from tensorflow import keras
```

```python
from keras import Sequential
from keras.layers import Dense
from keras.layers import Flatten
```

```python
(X_train.shape[1],)
```

```python
model=Sequential(
    [
        Flatten(input_shape=(30,)),
        Dense(20,activation='relu'),
        Dense(2,activation='sigmoid')
    ]
)
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(X_train_std,y_train,epochs=20,validation_split=0.1)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','val'],loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','val'],loc='upper left')
plt.show()
```

```python
loss,accuracy=model.evaluate(X_test_std,y_test)
```
